```python
import pandas as pd
import semopy
```

```python
df = pd.read_excel('SPSS_ThesisTh.xlsx')
```

```python
df.head()
```

```python
stress = ['S' + str(x) for x in range(1,11)]
burnout = ['BO' +str(x) for x in range(1,11)]
resilience = ['R' + str(x) for x in range(1,7)]
social_support = ['SO' + str(x) for x in range(1,13)]
```

```python
model = "# Measure\n"
model += f"stress =~ {' + '.join(stress)}\n" +\
    f"burnout =~ {' + '.join(burnout)}\n" +\
    f"resilience =~ {' + '.join(resilience)}\n" + \
    f"social_support =~ {' + '.join(social_support)}\n"
model += "# Regression\n"
model += "resilience ~ stress\n" +\
    "social_support ~ stress\n" +\
    "burnout ~ stress + resilience + social_support"
print(model)
```

```python
def fit(model, df):
    mod = semopy.Model(model)
    res = mod.fit(df)
    return mod.inspect()
```

```python
fit(model, df)
```
